```python
import streamlit as st
import matplotlib.pyplot as plt
import pandas as pd

# Load hasil prediksi dari DataFrame (df_test)
df_test = pd.read_csv('Hasil-Prediksi.csv')  

# Visualisasi grafik
st.title('Grafik Hasil Prediksi')
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df_test['Close'], label='Actual Close Price', color='blue')
ax.plot(df_test['MA'], label='Predicted Close Price', color='green', linestyle='dashed')
ax.set_title('Actual vs. Predicted Close Price')
ax.set_xlabel('Date')
ax.set_ylabel('Close Price')
ax.legend()

# Tampilkan grafik menggunakan Streamlit
st.pyplot(fig)
```
